# Supplementary Figure S6 — Merged (MA + DBS) brain region participation maps

Region participation maps built from n-plets that are high-performing **across both**
datasets simultaneously (merged tail). These maps reveal brain regions consistently
participating in high-order interactions that distinguish consciousness regardless of
the specific paradigm.

**Input**: `results/R1_C_region_maps_PRAUC_deltaO.pkl.gz`  
Requires a `'MA+DBS'` key produced when `R1_B_region_sampling.ipynb` was run with
the combined-dataset option. If absent, consult `S4_combined_dataset.ipynb` or
re-run `R1_B` with `DATASET = 'MA+DBS'`.

```python
from pathlib import Path
import os

def ensure_project_root(target_name: str = "high-order-anesthesia") -> Path:
    cwd = Path.cwd().resolve()
    if cwd.name == target_name:
        return cwd
    for parent in cwd.parents:
        if parent.name == target_name:
            os.chdir(parent)
            return parent
    raise RuntimeError(f"Could not find '{target_name}' in path.")

ROOT = ensure_project_root()
print(f"Now in: {ROOT.name}")
```

```python
import pickle
import gzip
import plotly.io as pio
pio.renderers.default = "notebook_connected"

from src.hoi_anesthesia.plotting import plot_cocomac_region_values
```

```python
maps_path = "results/R1_C_region_maps_PRAUC_deltaO.pkl.gz"
with gzip.open(maps_path, "rb") as fh:
    region_maps = pickle.load(fh)

available_keys = list(region_maps.keys())
print("Available dataset keys:", available_keys)

if "MA+DBS" not in available_keys:
    print("WARNING: 'MA+DBS' key not found. Re-run R1_B with combined-dataset option.")
```

## Merged maps at each order

```python
cmap_by_polarity = {"c_gt_nr": "Blues", "nr_gt_c": "Reds"}
variable = "region_counts_percent"

for order in sorted(region_maps["MA+DBS"].keys()):
    for polarity in ["c_gt_nr", "nr_gt_c"]:
        if polarity not in region_maps["MA+DBS"][order]:
            continue
        z_map = region_maps["MA+DBS"][order][polarity][variable]
        pol_label = (r"Ω_C > Ω_NR" if polarity == "c_gt_nr" else r"Ω_NR > Ω_C")
        print(f"\nMA+DBS | {pol_label} | order {order}")
        fig = plot_cocomac_region_values(
            z_map,
            cmap_name=cmap_by_polarity[polarity],
            colorbar_title="% of n-plets",
            fontsize=20,
            height=400,
            width=800,
            distance=0.73,
        )
        fig.show()
```
